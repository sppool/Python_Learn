```python
import numpy as np
from PIL import Image
```

```python
pp = Image.open('dog.png')
ar = np.array(pp)
up = 230
down = 150
arr = (((ar[:,:,2] < up).astype(int) + (ar[:,:,2] > down ).astype(int)) != 2).astype(int)
for i in range(3):
    ar[:,:,i] = ar[:,:,i]*arr

for i in range(3):
    ar[:,:,i] = ar[:,:,i] + ((arr == 0)*255)
ss = Image.fromarray(ar)
ss
```

```python
ss.save('flask dog.png')
```
