```python
import numpy as np
from PIL import Image, ImageFilter
```

```python
# 顯示圖片
def show_(arr):
    return Image.fromarray(np.uint8(arr))
```

```python
p = Image.open("Fish.jpg").resize((900, 600))
arr = np.array(p)
ar = arr[:,:,0]*0.9 + arr[:,:,1]*0.05 + arr[:,:,2]*0.05
val = 150
tmp = (ar < val)
ar2 = (ar * tmp) - ((tmp-1)*255)
show_(ar2)
```

```python
pp = Image.open('dog.png').resize((900, 600))
ar = np.array(pp)
up = 230
down = 150
arr = (((ar[:,:,2] < up).astype(int) + (ar[:,:,2] > down ).astype(int)) != 2).astype(int)
for i in range(3):
    ar[:,:,i] = ar[:,:,i]*arr

for i in range(3):
    ar[:,:,i] = ar[:,:,i] + ((arr == 0)*255)
ar = ar.mean(2)
show_(ar)
```

```python
n = 5
c = (ar - ar2) / n
show_(ar2 + (c * 0))
```
