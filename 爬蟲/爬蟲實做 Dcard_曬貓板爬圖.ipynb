````python
headers = {'user-agent':'Mozilla/5.0 (X11; Linux x86_64)'}
def get_soup(url, headers=headers):
    r = requests.get(url, headers=headers)
    rt = r.text
    soup = bs4.BeautifulSoup(rt, "html.parser")
    
    return soup
````

````python
url = 'https://www.dcard.tw/f/show_cats'

soup = get_soup(url)
````

#### 變成連結了

````python
server_url = url[:url.find('.tw') + 3] 
a_lst = []
for a in soup.select('div.PostList_entry_1rq5Lf a.PostEntry_root_V6g0rd'):
    a_url = server_url + a['href']
    a_lst.append(a_url)
    print(a_url)
````
