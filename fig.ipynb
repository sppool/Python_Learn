```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('classic')  # 風格設定
# plt.rcParams['font.sans-serif'] = ['Gen Jyuu Gothic P']  # 中文支援
```

```python
n = 750
s = np.abs(np.random.normal(4, 2, size=n))
x1 = np.random.normal(10, 2.5, n)
y1 = np.random.normal(10, 5, n)
x2 = np.random.normal(10, 5, n)
y2 = np.random.normal(-10, 2.5, n)
x3 = np.random.normal(-10, 5, n)
y3 = np.random.normal(10, 2.5, n)
x4 = np.random.normal(-10, 2.5, n)
y4 = np.random.normal(-10, 5, n)
data = pd.DataFrame({'x1':x1, 'y1':y1, 'x2':x2, 'y2':y2, 'x3':x3, 'y3':y3, 'x4':x4, 'y4':y4})
```

```python
s = np.linspace(2, 10, n)*12
fig = plt.figure(facecolor='#ccddee', figsize=(10, 7.5))
ax = plt.axes((0.1, 0.1, 0.8, 0.8))
col = data.columns
n = len(col)//2
b1 = True
b2 = True
colors = matplotlib.cm.gist_rainbow(np.linspace(0, 1, n))
alpha = 0.65
markers = ['o','^','p','s']
labels = ['A', 'B', 'C', 'D']
for i in range(n):
        x = data[col[(2 * i)]]
        y = data[col[(2 * i) + 1]]
        ax.scatter(x, y, edgecolors='k', marker=markers[i], s=s, label=labels[i], color=colors[i], alpha=alpha, )
ax.axis((-30, 30, -30, 30))
ax.set_title('$Scatter$')
ax.set_xlabel('$(mm)$')
ax.set_ylabel('$(mm)$')
plt.legend()
ax.grid()
# fig.savefig('scatter.png', facecolor='#ccddee')
```

```python
siz = 2000
x1 = np.random.normal(100, 10, siz)
x2 = np.random.uniform(60, 120, siz)
x3 = np.random.logistic(65, 2.5, siz)
df = pd.DataFrame({'data1':x1, 'data2':x2, 'data3':x3})
col = df.columns
```

```python
fig = plt.figure(facecolor='#ccddee', figsize=(10, 7.5))
ax = plt.axes((0.1, 0.1, 0.8, 0.8))
minn = df.values.min()
maxx = df.values.max()
colors = matplotlib.cm.brg(np.linspace(0, 1, 3))
labels = ['normal distribution', 'uniform distribution', 'logistic distribution']
for i in range(len(col)):
    ax.hist(df[col[i]], bins=40, color=colors[i], range=(minn, maxx), ec=None, label=labels[i], alpha=0.65)
ax.set_title('$distribution$')
ax.set_xlabel('$x$')
ax.set_ylabel('$count$')
ax.legend()
# plt.box()
ax.grid()
# fig.savefig('his.png', facecolor='#ccddee')
```

```python
fig = plt.figure(facecolor='#ccddee', figsize=(10, 7.5))
ax = plt.axes((0, 0, 0.75, 0.9))

col = df.columns
labels = df[col[0]].values
sizes = df[col[1]].values
colors = matplotlib.cm.gist_rainbow(np.linspace(0, 1, 8))

explode = np.linspace(0, 0.2, 8)
fig.set_facecolor('#abcdef')
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=15)
plt.title('$pets-pie$')
fig.legend(loc=7)
# fig.savefig('pie.png', facecolor='#ccddee')
```
